## 21st Century Races

```python
races_by_year = races_df.groupby('year')['raceId'].apply(list).reset_index(name='race_list')

races_by_year['num_races'] = races_by_year['race_list'].str.len()

races_by_year.year.min()
```

```python
unnested_races = races_by_year.explode('race_list')

#create dict to reference year of race
race_year_dict = dict(zip(unnested_races.race_list, unnested_races.year))
```

## Adding Year to Constructor Results

```python
#First, narrow down to results after 2000
results_df = pd.read_csv('data/Formula1/constructor_results.csv')
with_years = results_df.copy()
```

```python
with_years['year'] = with_years['raceId'].map(race_year_dict)
with_years.groupby(['year','constructorId']).count()
```

## Extract Constructor Names

```python
constructors = pd.read_csv('data/Formula1/constructors.csv')
constructors
```

```python
constructor_names = dict(zip(constructors.constructorId, constructors.name))
```

## All Years

```python
allYears = with_years.copy().reset_index()
allYears = allYears

allYears['constructorName'] = allYears['constructorId'].map(constructor_names)
allYears
```

## 2000s+ Only

```python
twentyfirst = with_years[with_years['year'].between(2000,2022)].copy().reset_index()

twentyfirst.sort_values('points')
```

## Constructor Points Per Year (2000-2022)

```python
with_cnames = twentyfirst.copy()
with_cnames['constructorName'] = with_cnames['constructorId'].map(constructor_names)
with_cnames
```

```python
constructor_pts = pd.DataFrame(with_cnames.groupby(['year','constructorId']).sum('points').reset_index())
```

```python
constructor_pts_year = with_years[with_years['year'].between(2000,2022)].copy()

constructor_pts_year = pd.DataFrame(constructor_pts_year.groupby(['year','constructorId']).sum('points').
                                    reset_index())

constructor_pts_year['constructorName'] = constructor_pts_year['constructorId'].map(constructor_names)
```

```python
constructor_pts_year['year'] = constructor_pts_year['year'].astype('category')
constructor_pts_year["rank"] = constructor_pts_year.groupby(['year'])["points"].rank(method="dense", ascending=False)
constructor_pts_year["rank"] = constructor_pts_year['rank'].astype('int')


constructor_pts_year['pointsPct']= (constructor_pts_year['points'] / constructor_pts_year.groupby('year')['points'].transform('sum') * 100)
constructor_pts_year['pointsPct']= constructor_pts_year['pointsPct']

constructor_pts_year = constructor_pts_year.sort_values(['year','points'], ascending=False)

#constructor_pts_year.to_csv('transformed_data/const_pts_by_year.csv')
```

## Points By Year (ALL)

```python
all_pts = allYears.copy()
all_pts
```

```python
allYears_pts = pd.DataFrame(allYears.groupby(['year','constructorId','constructorName']).sum('points').reset_index())
allYears_pts
```

```python
allYears_pts['year'] = allYears_pts['year'].astype('category')
allYears_pts["rank"] = allYears_pts.groupby(['year'])["points"].rank(method="dense", ascending=False)
allYears_pts["rank"] = allYears_pts['rank'].astype('int')


allYears_pts['pointsPct']= (allYears_pts['points'] / allYears_pts.groupby('year')['points'].transform('sum') * 100)
allYears_pts['pointsPct']= allYears_pts['pointsPct']

allYears_pts = allYears_pts.sort_values(['year','points'], ascending=False).reset_index(drop=True)

#allYears_pts.to_csv('transformed_data/allYears_pts.csv')
```

```python
allYears_pts
```

```python
top3_alltime = allYears_pts[allYears_pts['rank']<=3]
top3_alltime = allYears_pts[allYears_pts['rank']<=3]
top3_alltime
```

```python
all_time_top3 = pd.pivot(top3_alltime, index='year', columns='constructorName', values='pointsPct').fillna(0).reset_index()

all_time_top3

#all_time_top3.to_csv('transformed_data/all_time_top3.csv', index=False)
```
